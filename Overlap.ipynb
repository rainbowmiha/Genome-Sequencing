```
# Create your dictionary class 
class my_dictionary(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value 
```

```
# A program which finds which patterns (in a given list of dna seqs (strings))
# and finds their adjacent (overlapping patterns)

def overlap(patterns):

  # sort out patterns lexicographically  
  patterns.sort()

  # using list comprehension to remove duplicated from list  
  res = [] 
  [res.append(x) for x in patterns if x not in res]
  patterns = res

  #length of patterns
  k = len(patterns[0])

  # declare a dictionary, using the class declared above
  adj_lists = my_dictionary() 

  # an list to temporarily store the patterns adjacent to pattern i  
  patt_adj_to_i = []

  # a list to store the patterns with 0 adj patterns
  zero_adj_list = []

  # loops over patterns and finds their adjacent patterns  
  for i in patterns:
    for j in patterns:
      if i[1:k] == j[0:k-1]:
        patt_adj_to_i.append(j)

    # add list of patterns adjacent to i     
    adj_lists[i] = patt_adj_to_i

   # if list of adjacent patterns is empty, take note  
    if len(patt_adj_to_i)==0:
      zero_adj_list.append(i)

    # empty patt_adj_to_i
    patt_adj_to_i = []  

  for x in zero_adj_list:
    del adj_lists[x]    


  for key in adj_lists.keys():
    print (str(key) + ' -> ' + ",".join(adj_lists[key])) 
  
```
