```
# a program which takes in a dna seq (string) and a k-mer length(k - int)
# and generates the non-overlapping kmers of the dna seq
def compositionk(dnaseq,k):
  kmers = []
  for i in range(len(dnaseq)-k+1):
    kmers.append(dnaseq[i:i+k])
  kmers.sort()
  for x in kmers:
    print (x)
 
```
